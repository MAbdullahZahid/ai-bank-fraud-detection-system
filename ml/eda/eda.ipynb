```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Paths


DATA_PATH = "../dataset/raw/train.csv"
OUTPUT_PATH = "plots"

os.makedirs(OUTPUT_PATH, exist_ok=True)


# Load Dataset


print("=" * 60)
print("Loading Dataset...")
print("=" * 60)

df = pd.read_csv(DATA_PATH)

print("\nDataset Loaded Successfully!")


# Dataset Shape


print("\n" + "=" * 60)
print("Dataset Shape")
print("=" * 60)

print(f"Rows    : {df.shape[0]:,}")
print(f"Columns : {df.shape[1]}")


# Data Types


print("\n" + "=" * 60)
print("Data Types")
print("=" * 60)

print(df.dtypes)


# Missing Values


print("\n" + "=" * 60)
print("Missing Values")
print("=" * 60)

missing = df.isnull().sum()
print(missing)


# Duplicate Rows


print("\n" + "=" * 60)
print("Duplicate Rows")
print("=" * 60)

duplicates = df.duplicated().sum()
print(f"Duplicate Rows : {duplicates:,}")


# Basic Statistics


print("\n" + "=" * 60)
print("Statistical Summary")
print("=" * 60)

print(df.describe())


# Target Distribution


print("\n" + "=" * 60)
print("Fraud Distribution")
print("=" * 60)

print(df["isFraud"].value_counts())

print("\nPercentage")

print(df["isFraud"].value_counts(normalize=True) * 100)


# Transaction Types


print("\n" + "=" * 60)
print("Transaction Types")
print("=" * 60)

print(df["type"].value_counts())


# Fraud by Transaction Type


print("\n" + "=" * 60)
print("Fraud by Transaction Type")
print("=" * 60)

fraud_type = pd.crosstab(df["type"], df["isFraud"])

print(fraud_type)


# Zero Balance Analysis


print("\n" + "=" * 60)
print("Zero Balance Analysis")
print("=" * 60)

print("oldbalanceOrg == 0 :", (df["oldbalanceOrg"] == 0).sum())

print("newbalanceOrig == 0 :", (df["newbalanceOrig"] == 0).sum())

print("oldbalanceDest == 0 :", (df["oldbalanceDest"] == 0).sum())

print("newbalanceDest == 0 :", (df["newbalanceDest"] == 0).sum())


# Amount Statistics


print("\n" + "=" * 60)
print("Transaction Amount Statistics")
print("=" * 60)

print(df["amount"].describe())


# Correlation Matrix


print("\nGenerating Correlation Matrix...")

numeric_df = df.select_dtypes(include=["number"])

plt.figure(figsize=(10, 8))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Matrix")

plt.tight_layout()

plt.savefig(f"{OUTPUT_PATH}/correlation_matrix.png")

plt.close()


# Fraud Distribution Plot


plt.figure(figsize=(6, 5))

sns.countplot(data=df, x="isFraud")

plt.title("Fraud Distribution")

plt.savefig(f"{OUTPUT_PATH}/fraud_distribution.png")

plt.close()


# Transaction Type Plot


plt.figure(figsize=(8, 5))

sns.countplot(data=df, x="type")

plt.title("Transaction Types")

plt.xticks(rotation=30)

plt.tight_layout()

plt.savefig(f"{OUTPUT_PATH}/transaction_types.png")

plt.close()


# Fraud by Type Plot


fraud_type.plot(
    kind="bar",
    figsize=(8, 5)
)

plt.title("Fraud by Transaction Type")

plt.ylabel("Count")

plt.tight_layout()

plt.savefig(f"{OUTPUT_PATH}/fraud_by_transaction_type.png")

plt.close()


# Amount Distribution


plt.figure(figsize=(8,5))

plt.hist(df["amount"], bins=100)

plt.title("Transaction Amount Distribution")

plt.xlabel("Amount")

plt.ylabel("Frequency")

plt.tight_layout()

plt.savefig(f"{OUTPUT_PATH}/amount_distribution.png")

plt.close()


# Boxplot


plt.figure(figsize=(8,5))

sns.boxplot(x=df["amount"])

plt.title("Transaction Amount Boxplot")

plt.tight_layout()

plt.savefig(f"{OUTPUT_PATH}/amount_boxplot.png")

plt.close()


# Fraud Percentage by Type


fraud_percent = pd.crosstab(
    df["type"],
    df["isFraud"],
    normalize="index"
) * 100

print("\n" + "=" * 60)
print("Fraud Percentage by Transaction Type")
print("=" * 60)

print(fraud_percent)


# Top 10 Largest Transactions


print("\n" + "=" * 60)
print("Top 10 Largest Transactions")
print("=" * 60)

print(
    df.sort_values(
        by="amount",
        ascending=False
    ).head(10)
)


# Dataset Memory Usage


print("\n" + "=" * 60)
print("Memory Usage")
print("=" * 60)

memory = df.memory_usage(deep=True).sum() / 1024**2

print(f"{memory:.2f} MB")


# Finished


print("\n" + "=" * 60)
print("EDA COMPLETED SUCCESSFULLY")
print("=" * 60)

print("\nGraphs saved inside:")

print("eda/plots/")
```
